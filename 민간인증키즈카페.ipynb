# 위도 경도 값을 찾기 위해 vworld api 이용 

```python
import requests
import pandas as pd
import urllib.parse
import time

# VWorld API 키 (자신의 VWorld API 키로 변경)
VWORLD_API_KEY = '47171D39-1C3B-3F42-9575-F23537C84C1C'

# VWorld API를 통해 주소 -> 위경도 변환 함수
def get_coords_from_vworld(address):
    base_url = "http://api.vworld.kr/req/address"
    params = {
        "service": "address",
        "request": "getcoord",
        "version": "2.0",
        "crs": "epsg:4326",  # WGS84 좌표계
        "address": address,
        "refine": "true",  # 주소 정확도 높이기
        "simple": "false",  # 상세 주소 반환
        "format": "json",  # JSON 형식
        "key": VWORLD_API_KEY,  # API 키
    }
    
    try:
        query_string = urllib.parse.urlencode(params)
        response = requests.get(f"{base_url}?{query_string}")
        if response.status_code == 200:
            data = response.json()
            if data["response"]["status"] == "OK" and data["response"]["result"]["point"]:
                lat = data["response"]["result"]["point"]["y"]  # 위도
                lon = data["response"]["result"]["point"]["x"]  # 경도
                return lat, lon
    except Exception as e:
        print(f"VWorld API 오류: {e}")
    return None, None  # 주소에 대한 위경도 정보를 찾을 수 없는 경우

# 위도, 경도 값을 새 컬럼에 추가
df['위도'] = None
df['경도'] = None

# 각 행에 대해 주소로 위도/경도 가져오기
for idx, row in df.iterrows():
    address = row['도로명주소']  # 또는 '지번주소'를 사용할 수 있음
    lat, lon = get_coords_from_vworld(address)
    df.at[idx, '위도'] = lat
    df.at[idx, '경도'] = lon
    print(f"{address} → 위도: {lat}, 경도: {lon}")
    time.sleep(0.2)  # API 호출 제한을 피하기 위해 잠시 대기

# 결과 데이터프레임 확인
df.head()

# 최종 데이터 CSV로 저장
output_path = r'C:\Users\user\Desktop\디지털 마스터\프로젝트\서울형 인증제 키즈카페_위경도.csv'
df.to_csv(output_path, index=False, encoding='utf-8-sig')
print("위경도 추가 완료, 파일 저장됨.")
```

```python
import requests
import pandas as pd
import time

# 카카오맵 API 키
KAKAO_API_KEY = '49f6629ef3cc7efb7fe16914e0e16ced'

# 데이터 불러오기
df = pd.read_csv('./dataset/서울형 인증제 키즈카페 주소.csv', encoding='cp949')

# 함수 : 카카오맵 API로 주소를 위경도 변환하기
def get_coords_from_kakao(address):
    url = "https://dapi.kakao.com/v2/local/search/address.json"
    headers = {"Authorization": f"KakaoAK {KAKAO_API_KEY}"}
    params = {"query": address}
    
    response = requests.get(url, headers=headers, params=params)
    if response.status_code == 200:
        data = response.json()
        if data['documents']:
            lat = data['documents'][0]['y']  
            lon = data['documents'][0]['x'] 
            return lat, lon
    return None, None  # 위경도 정보 없는 경우

# 새 칼럼 만들어서 위경도 추가
df['위도'] = None
df['경도'] = None

# 각 행에 대해 주소로 위도/경도 가져오기
for idx, row in df.iterrows():
    address = row['도로명주소']  # 또는 '지번주소'를 사용할 수 있음
    lat, lon = get_coords_from_kakao(address)
    df.at[idx, '위도'] = lat
    df.at[idx, '경도'] = lon
    print(f"{address} → 위도: {lat}, 경도: {lon}")
    time.sleep(0.2)  # API 호출 제한을 피하기 위해 잠시 대기

# 결과 데이터프레임 확인
df.head()

# 최종 데이터 CSV로 저장
output_path = r'C:\Users\user\Desktop\디지털 마스터\프로젝트\서울형 인증제 키즈카페_위경도.csv'
df.to_csv(output_path, index=False, encoding='utf-8-sig')
print("위경도 추가 완료, 파일 저장됨.")
```

# API로 행정동 불러오기

```python
import pandas as pd
import requests
from tqdm import tqdm

df = pd.read_csv('./dataset/서울형 인증제 키즈카페_위경도_최종.csv')

# 카카오 API 설정
KAKAO_API_KEY = '49f6629ef3cc7efb7fe16914e0e16ced'  # ← 여기에 본인의 카카오 REST API 키 입력
headers = {"Authorization": f"KakaoAK {KAKAO_API_KEY}"}

# 행정동명을 도로명주소로부터 추출하는 함수
def get_admin_dong_from_road_address(road_address):
    try:
        url = "https://dapi.kakao.com/v2/local/search/address.json"
        params = {"query": road_address}
        response = requests.get(url, headers=headers, params=params)
        if response.status_code == 200:
            documents = response.json().get('documents')
            if documents:
                address = documents[0]['address']
                return address.get('region_3depth_name')  # 행정동 이름
        return None
    except Exception as e:
        print(f"Error at {road_address}: {e}")
        return None

# tqdm을 사용해 진행 상황 시각화
tqdm.pandas()

# 도로명주소 → 행정동명 추출
df['행정동'] = df['도로명주소'].progress_apply(get_admin_dong_from_road_address)

# 결과 확인
print(df[['도로명주소', '행정동']].head())

# 결과 저장
output_path = r'C:\Users\user\Desktop\디지털 마스터\프로젝트\서울형 인증제 키즈카페_위경도_행정동.csv'
df.to_csv(output_path, index=False, encoding='utf-8-sig')
print("완료!")
```

# API로 행정동 코드 불러오기

```python
import pandas as pd
import requests
from tqdm import tqdm

# 데이터 불러오기
df = pd.read_csv('./dataset/서울형 인증제 키즈카페_위경도_행정동.csv')

# 카카오 API 설정
KAKAO_API_KEY = '49f6629ef3cc7efb7fe16914e0e16ced'  # ← 본인의 키로 바꿔주세요
headers = {"Authorization": f"KakaoAK {KAKAO_API_KEY}"}

# 위도, 경도 → 행정동명 & 코드 추출 함수
def get_admin_dong_info(lat, lon):
    try:
        url = "https://dapi.kakao.com/v2/local/geo/coord2regioncode.json"
        params = {"x": lon, "y": lat}
        response = requests.get(url, headers=headers, params=params)
        if response.status_code == 200:
            documents = response.json().get('documents', [])
            for doc in documents:
                if doc['region_type'] == 'H':  # 'H'는 행정동, 'B'는 법정동
                    return pd.Series([doc['region_3depth_name'], doc['code']])
        return pd.Series([None, None])
    except Exception as e:
        print(f"Error at ({lat}, {lon}): {e}")
        return pd.Series([None, None])

# 진행 표시
tqdm.pandas()

# 위도/경도 기준 행정동명과 코드 추출
df[['행정동', 'api행정동코드']] = df.progress_apply(
    lambda row: get_admin_dong_info(row['위도'], row['경도']), axis=1
)

print(df)
```

```python
# 결과 저장
output_path = r'C:\Users\user\Desktop\디지털 마스터\프로젝트\서울형 인증제 키즈카페_행정동_코드추가.csv'
df.to_csv(output_path, index=False, encoding='utf-8-sig')
print("완료!")
```
